Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing Libraries required for deep Learning models

```python
import tensorflow as tf
from tensorflow import keras
```

Importing the layers

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import InputLayer, Dense
```

Importing the Dataset

```python
Train_Data = pd.read_csv('Train_Cleaned.csv')
Test_Data = pd.read_csv('Test_Cleaned.csv')
```

```python
#Viewing the data
Train_Data.head()
```

```python
#Scaling year column
#Identifying the minimum and maximum year of training data
min_year = Train_Data['Outlet_Establishment_Year'].min()
max_year = Train_Data['Outlet_Establishment_Year'].max()
print(min_year)
print(max_year)
```

```python
Train_Data['Outlet_Establishment_Year'] = (Train_Data['Outlet_Establishment_Year']-min_year)/(max_year - min_year)
```

```python
#Viewing the dataset now
Train_Data.head()
```

```python
Test_Data.head()
```

```python
#Scaling the year of test data
Test_Data['Outlet_Establishment_Year'] = (Test_Data['Outlet_Establishment_Year']-min_year)/(max_year - min_year)
```

```python
Test_Data.head()
```

Dropping the first column since its not wanted

```python
Train_Data = Train_Data.drop(Train_Data.columns[0], axis=1)
```

```python
Train_Data.head()
```

```python
Test_Data = Test_Data.drop(Test_Data.columns[0], axis=1)
```

```python
Test_Data.head()
```

Splitting the independent and Dependent Variables

```python
X = Train_Data.drop('Item_Outlet_Sales',axis=1)
y = Train_Data['Item_Outlet_Sales']
```

Seeing the Shape of independent and dependent variables

```python
X.shape, y.shape
```

Seperating Training and Validation

```python
#Importing the libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Seperating training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50, test_size=0.2)
```

```python
#Checking the Shape
(X_train.shape, y_train.shape),(X_test.shape,y_test.shape)
```

```python
#Defining the input shape of the model
model_input_shape = X_train.shape[1]
print(model_input_shape)
```

Building the Model

```python
model = Sequential()
```

```python
model.add(InputLayer(input_shape=(model_input_shape,)))
```

```python
#Adding the Hidden Layers
model.add(Dense(units = 20, activation ='relu'))
model.add(Dense(units = 15, activation = 'relu'))
model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 5, activation = 'relu'))
model.add(Dense(units = 1, activation = 'linear'))  #Output neuron
```

summary of the model

```python
model.summary()
```

Compiling the model

```python
Test_Data.isna().sum()
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adagrad(learning_rate=0.01), metrics=['mse'])
```

```python
#Training the model
model_history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=60)
```
