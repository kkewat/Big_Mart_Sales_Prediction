```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('test.csv')
```

```python
#Viewing the null Values
data.isnull().sum()
```

```python
#Viewing the data type of the columns
data.dtypes
```

Filling the Null Values

```python
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Weight'].mean())
data['Outlet_Size'] = data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0])
```

```python
#Checking if there is any remaining null values
data.isna().sum()
```

```python
#Viewing the Data
data.head()
```

Encoding the Data

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].map({'LF':1,'Low Fat':1,'low fat':1,'Low fat':1,'low Fat':1,'lf':1,'regular':2,'Regular':2,'reg':2,'Reg':2})
data['Outlet_Size'] = data['Outlet_Size'].map({'High':1,'Medium':2,'Small':3})
data['Outlet_Location_Type'] = data['Outlet_Location_Type'].map({'Tier 1':1,'Tier 2':2,'Tier 3':3})
data['Outlet_Type'] = data['Outlet_Type'].map({'Grocery Store':1,'Supermarket Type1':2,'Supermarket Type2':3,'Supermarket Type3':4})
```

```python
data = pd.get_dummies(data, columns=['Item_Type','Outlet_Identifier'])
```

```python
data.head()
```

```python
#Dropping the columns which haves all the unique variable
data = data.drop(columns='Item_Identifier')
```

```python
data.head()
```

Scaling the data

```python
#Filling the min and max of the columns we found in training dataset so that the dataset is even
item_weight_min = 4.555
item_weight_max = 21.35

item_mrp_min = 31.29
item_mrp_max = 266.8884

item_outlet_sales_min = 33.29
item_outlet_sales_max = 13086.9648
```

```python
#Scalling the Data
# Min MAx Scaling
data['Item_Weight'] = (data['Item_Weight']-item_weight_min)/(item_weight_max-item_weight_min)
data['Item_MRP'] = (data['Item_MRP']-item_mrp_min)/(item_mrp_max-item_mrp_min)
```

```python
#Viewing our data
data.head()
```

```python
#Exporting our test cleaned dataset
data.to_csv('Test_Cleaned.csv')
```
