```python
#importing the required datasets
import pandas as pd
import numpy as np
```

Importing the dataset

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
#Viewing the Null Values in the Dataset
data.isna().sum()
```

```python
#Viewing the data type of the columns
data.dtypes
```

#Filling the null values

```python
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Weight'].mean())
```

```python
data['Outlet_Size'] = data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0])
```

```python
#Checking if there is any remaining null values
data.isna().sum()
```

```python
#Viewing the Data
data.head()
```

ENcoding the Data

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].map({'LF':1,'Low Fat':1,'Regular':2,'reg':2})
```

```python
data.head()
```

```python
data['Outlet_Size'] = data['Outlet_Size'].map({'High':1,'Medium':2,'Small':3})
data['Outlet_Location_Type'] = data['Outlet_Location_Type'].map({'Tier 1':1,'Tier 2':2,'Tier 3':3})
data['Outlet_Type'] = data['Outlet_Type'].map({'Grocery Store':1,'Supermarket Type1':2,'Supermarket Type2':3,'Supermarket Type3':4})
```

```python
data = pd.get_dummies(data, columns=['Item_Type','Outlet_Identifier'])
```

```python
data.head()
```

```python
#Dropping the columns which haves all the unique variable
data = data.drop(columns='Item_Identifier')
```

```python
data.head()
```

## **SCaling the Data**

```python
#FInding the min and max of the columns to be scaled
item_weight_min = data['Item_Weight'].min()
item_weight_max = data['Item_Weight'].max()

item_mrp_min = data['Item_MRP'].min()
item_mrp_max = data['Item_MRP'].max()

item_outlet_sales_min = data['Item_Outlet_Sales'].min()
item_outlet_sales_max = data['Item_Outlet_Sales'].max()
```

```python
#Printing the minimum and maximum
print('Minimun_Weight ',item_weight_min,' Maximum Weight ',item_weight_max)
print('Minimum Mrp ', item_mrp_min,' Maximum Mrp ',item_mrp_max)
print('Minimum Outlet sales ',item_outlet_sales_min, ' Maximum Sales', item_outlet_sales_max)
```

```python
# Min MAx Scaling
data['Item_Weight'] = (data['Item_Weight']-item_weight_min)/(item_weight_max-item_weight_min)
data['Item_MRP'] = (data['Item_MRP']-item_mrp_min)/(item_mrp_max-item_mrp_min)
data['Item_Outlet_Sales'] = (data['Item_Outlet_Sales']-item_outlet_sales_min)/(item_outlet_sales_max-item_outlet_sales_min)
```

```python
data.head()
```

```python
data.to_csv('Train_Cleaned.csv')
```
